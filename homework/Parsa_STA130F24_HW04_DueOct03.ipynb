````python
# We are working with a dataset about patient's health score before and after some treatment. 
# We want to find out whether the treatment is effective.
# The null hypothesis is no effect means we hypothesize this treatment does not work, aka the health score before and after treatment will be the same.

import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/Vizxe/sta130/refs/heads/main/homework/alitech.csv")
# Rename for convenience 
df = df.rename(columns={"InitialHealthScore":"init", "FinalHealthScore": "final"})
# Remove uncessary column
del df["PatientID"]
df
````

````python
# let's view the patient's initial health score and also their health score after the treatement

import plotly.express as pe
pe.histogram(df, x="init", nbins=5).show(renderer="png")
pe.histogram(df, x="final", nbins=5).show(renderer="png")
````

````python
# And let's see if the ages are not biased
pe.histogram(df, x="Age", nbins=5).show(renderer="png")
````

````python
# But what we really care about is the difference in health scores.
df['diff'] = df.final - df.init
pe.histogram(df, x="diff", nbins=10).show(renderer="png")
````

````python
import numpy as np
np.random.seed(69)

# I'm going to do bootstrapping on the mean of diff.
means = []
n = len(df['diff'])
for i in range(10000):
    dist = df['diff'].sample(n=n,replace=True)
    means.append(dist.mean())
````

````python
# Let's visualize bootstrapping on the mean 

fig = pe.histogram(means, nbins=100)
fig.show(renderer="png")
````

````python
# Looks normal! 
# See the 95% CI
import plotly.graph_objects as go

fig.add_vline(x=np.percentile(means, 2.5), line_dash="dash", annotation_text="2.5%")
fig.add_vline(x=np.percentile(means, 97.5), line_dash="dash", annotation_text="97.5%")
fig.show(renderer="png")
````

````python
# Since 0 is not within the range of the 95% CI, we conclude that we are 95% sure the treatment has "some effect", 
# Thus we reject the null hypothesis, and propose H_\alpha, which says the the treatement does positively impact the patient to some extent.
````
